```
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
```

```
# Binomial Dist
n = 10 # No. of Observation
p = 0.5 # Probability
k = np.arange(1,11) # Trials
stats.binom.pmf(k, n, p)
```

```
plt.plot(k,stats.binom.pmf(k, n, p))
plt.show()
```

```
# Normal Dist
mean = 0
sdiv = 1
x = np.linspace(-10,10,100)
plt.plot(x, stats.norm.pdf(x, mean, sdiv))
plt.show()
```

```
# Poisson Dist
rate = 10
n = np.arange(0,11)
stats.poisson.pmf(n, rate)
```
